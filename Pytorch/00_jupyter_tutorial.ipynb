```python
from tqdm.notebook import tqdm
```

```python
import time

# 반복문이 20번 실행
for i in tqdm(range(20)):
    time.sleep(.5) # 각 반복마다 0.5초 동안 대기 => 반복문이 한 번 돌때마다 0.5초의 지연 발생
```

- import time: time 모듈에는 시간을 다루는 함수들이 포함되어 있음
- tqdm 라이브러리: 진행 상태를 시각적으로 표시하는 라이브러리, 프로그래스 바(Progress bar)를 제공



## Magic

- %%time은 Jupyter Notebook에서 사용되는 매직 커맨드(Magic command) 중 하나로, 코드 셀의 실행 시간을 측정할 때 사용

```python
%%time
for i in tqdm(range(20)):
    time.sleep(.5)

# 해당 셀의 전체 실행 시간 / 시작부터 끝까지 걸린 시간 / Wall time + CPU time 출력
```

- Wall time: 실제 경과된 시간, 즉 시계로 측정하는 시간
- CPU time: CPU가 해당 셀을 실행하는 데 걸린 순수한 프로세싱 시간 (입출력(I/O) 작업이나 다른 대기 시간은 포함되지 않음)

```python
%%timeit
sum(x**2 for x in range(1,100000,10))
```

### %%timeit은 해당 셀의 코드를 여러 번 반복 실행하여 평균 실행 시간과 표준 편차를 측정
- 여러 번 실행하여 평균 시간을 계산합니다. 주로 짧은 코드의 정확한 실행 시간을 평가할 때 유용
