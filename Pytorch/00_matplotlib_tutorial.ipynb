```python
from faker import Faker
import pandas as pd
```

```python
# Faker 인스턴스 초기화
fake = Faker()
```

### Faker: 가짜 데이터(fake data)를 생성하는 데 사용되는 Python 라이브러리이다. 주로 테스트용 데이터나 샘플 데이터 만들 때 유용
- 실제 데이터가 필요하지 않은 상황에서 임의의 이름, 주소, 전화번호, 이메일, 날짜, 텍스트 등의 정보를 손쉽게 생성 가능
- Faker 라이브러리: 무작위로 생성된 데이터를 반환

```python
fake.name(), fake.random_int(min=18, max=80), fake.address()
```

```python
# 가짜 데이터 생성 (예: 100명의 이름과 나이)
data = {'Name': [fake.name() for _ in range(100)],
        'Age': [fake.random_int(min=18, max=80) for _ in range(100)]}

# Pandas DataFrame으로 변환
df = pd.DataFrame(data)
```

```python
df
```

```python
# 데이터 시각화 (나이 분포 히스토그램)
plt.figure(figsize=(10, 6))
plt.hist(df['Age'], bins=np.arange(18, 81, 1), color='skyblue', edgecolor='black')
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.grid(axis='y', alpha=0.75)
plt.show()
```
